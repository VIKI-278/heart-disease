# Data description


    1) Age: age of the patient [years]
    2) Sex: sex of the patient [M: Male, F: Female]
    3) ChestPainType: chest pain type [TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic]
    4) RestingBP: resting blood pressure [mm Hg]
    5) Cholesterol: serum cholesterol [mm/dl]
    6) FastingBS: fasting blood sugar [1: if FastingBS > 120 mg/dl, 0: otherwise]
    7) RestingECG: resting electrocardiogram results 
        [Normal: Normal, ST: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV),        LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria]
    8) MaxHR: maximum heart rate achieved [Numeric value between 60 and 202]
    9) ExerciseAngina: exercise-induced angina [Y: Yes, N: No]
    10) Oldpeak: oldpeak = ST [Numeric value measured in depression]
    11) ST_Slope: the slope of the peak exercise ST segment [Up: upsloping, Flat: flat, Down: downsloping]
    12) HeartDisease: output class [1: heart disease, 0: Normal]

## Encoding

```python
#identifying categorical variable
cat_variables=[]
for column in df.columns:
    if df[column].dtype == 'O':
        cat_variables.append(column)
```

```python
#Displaying categories of each cat var
for var in cat_variables:
    print(var,":",df[var].unique())
```

```python
# one hot encoding
one_hot_encoded = pd.get_dummies(df[['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']],drop_first=True)
```

```python
#concat
df_encoded = pd.concat([df, one_hot_encoded], axis=1)
```

```python
df_encoded.head()
```

```python
# removing original cat vars
df_encoded.drop(columns = cat_variables, inplace = True)
```

# Train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
X = df_encoded.drop(columns='HeartDisease')
y = df_encoded['HeartDisease']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

# Modeling

# We shall try out:  
1. Logistic regression
2. Naive Bayes classification
3. KNN classification and
4. Random Forest classifier and select the best model and try to tune them


# Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression

#base model training
lr = LogisticRegression()
LR_model = lr.fit(X_train,y_train)

#predicting for test and train
LR_pred_train = LR_model.predict(X_train)
LR_pred_test = LR_model.predict(X_test)
LR_y_train_prob = LR_model.predict_proba(X_train)
LR_y_test_prob =  LR_model.predict_proba(X_test)


fig, axes = plt.subplots(3, 1, figsize=(7, 15))

print('Logistic Regression base model:')

#printing accuracy,classification report and confusion matrix for training data
print('Train_summary:')
print('Accuracy = ',metrics.accuracy_score(y_train,LR_pred_train))
print(metrics.classification_report(y_train,LR_pred_train))
sns.heatmap((metrics.confusion_matrix(y_train,LR_pred_train)),annot=True,fmt='.5g'
            ,cmap='Greys',ax=axes[0]);
axes[0].set_xlabel('Predicted');
axes[0].set_ylabel('Actuals',rotation=0);
axes[0].set_title('Train_summary')

#printing accuracy,classification report and confusion matrix for test data
print('Test_summary:')
print('Accuracy = ',metrics.accuracy_score(y_test,LR_pred_test))
print(metrics.classification_report(y_test,LR_pred_test))
sns.heatmap((metrics.confusion_matrix(y_test,LR_pred_test)),annot=True,fmt='.5g'
            ,cmap='Greys',ax=axes[1]);
axes[1].set_xlabel('Predicted');
axes[1].set_ylabel('Actuals',rotation=0);
axes[1].set_title('Test_summary:')

fpr_train, tpr_train, thresholds_train = roc_curve(y_train, LR_y_train_prob[:, 1])
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, LR_y_test_prob[:, 1])
axes[2].plot(fpr_train, tpr_train,color='blue',label = 'train')
axes[2].plot(fpr_test, tpr_test,color='green',label = 'test')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
axes[2].set_title('ROC Curve')
plt.tight_layout()
plt.show()

LR_auc_roc_train = roc_auc_score(y_train, LR_y_train_prob[:, 1])
LR_auc_roc_test = roc_auc_score(y_test, LR_y_test_prob[:, 1])

# print the AUC-ROC score
print('AUC-ROC train:', LR_auc_roc_train)
print('AUC-ROC test:', LR_auc_roc_test)
```

```python
# lets try gaussian naive bayes
from sklearn.naive_bayes import GaussianNB 

#base model building
GNB_model=GaussianNB()
GNB_model.fit(X_train,y_train)

#predicting for test and train
GNB_pred_train = GNB_model.predict(X_train)
GNB_pred_test = GNB_model.predict(X_test)
GNB_y_train_prob = GNB_model.predict_proba(X_train)
GNB_y_test_prob = GNB_model.predict_proba(X_test)


fig, axes = plt.subplots(3, 1, figsize=(7, 15))

print('GaussianNB base model:')

#printing accuracy,classification report and confusion matrix for training data
print('Train_summary:')
print('Accuracy = ',metrics.accuracy_score(y_train,GNB_pred_train))
print(metrics.classification_report(y_train,GNB_pred_train))
sns.heatmap((metrics.confusion_matrix(y_train,GNB_pred_train)),annot=True,fmt='.5g'
            ,cmap='Greys',ax=axes[0]);
axes[0].set_xlabel('Predicted');
axes[0].set_ylabel('Actuals',rotation=0);
axes[0].set_title('Train_summary')

#printing accuracy,classification report and confusion matrix for test data
print('Test_summary:')
print('Accuracy = ',metrics.accuracy_score(y_test,GNB_pred_test))
print(metrics.classification_report(y_test,GNB_pred_test))
sns.heatmap((metrics.confusion_matrix(y_test,GNB_pred_test)),annot=True,fmt='.5g'
            ,cmap='Greys',ax=axes[1]);
axes[1].set_xlabel('Predicted');
axes[1].set_ylabel('Actuals',rotation=0);
axes[1].set_title('Train_summary')

fpr_train, tpr_train, thresholds_train = roc_curve(y_train, GNB_y_train_prob[:, 1])
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, GNB_y_test_prob[:, 1])
axes[2].plot(fpr_train, tpr_train,color='blue',label = 'train')
axes[2].plot(fpr_test, tpr_test,color='green',label = 'test')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
axes[2].set_title('ROC Curve')
plt.tight_layout()
plt.show()

GNB_auc_roc_train = roc_auc_score(y_train, GNB_y_train_prob[:, 1])
GNB_auc_roc_test = roc_auc_score(y_test, GNB_y_test_prob[:, 1])

# print the AUC-ROC score
print('AUC-ROC train:', GNB_auc_roc_train)
print('AUC-ROC test:', GNB_auc_roc_test)
```
